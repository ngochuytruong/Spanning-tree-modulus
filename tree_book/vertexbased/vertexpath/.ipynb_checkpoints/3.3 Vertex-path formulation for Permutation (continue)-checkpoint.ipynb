```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import cdd
from modulus_tools import demo_graphs
from modulus_tools import algorithms as alg
from itertools import product
import cvxopt as cvx
from cvxopt.solvers import qp
from fractions import Fraction
import math
from itertools import combinations
```

```python
'list of all sequences'
def sub_lists(per):
    subs = []
    for i in range(0, len(per)+1):
        temp = [list(x) for x in combinations(per, i)]
        if len(temp)>0:
            subs.extend(temp)
    return subs
'list of all increasing sequences'
def good_sub_lists(per):
    B= []
    for l in sub_lists(per):
        
        if len(l)<= 1:
            B.append(l)
        else:
            passed = True
            for i in range(0,len(l)-1):
                if l[i]>=l[i+1]:
                    passed = False
            if  passed:
                B.append(l)
    return B
'list of all maximal increasing sequences -index '
def MISS(l): 
    miss = []
    for t  in  good_sub_lists(l):
        passed= True
        for s in  good_sub_lists(l):
            if all(x in s for x in t) == True:
                if all(x in t for x in s)== False:
                    passed = False
        if passed:
            miss.append(t)
    return miss
'Matrix usage of all maximal increasing sequences'
def MISSR(l):
    rows = []
    for s in MISS(l):
        row = len(l)*[0]
        for i in range(len(l)):
            if l[i] in s:
                row[i]=1        
        rows.append(row)
    return rows
'list of all maximal increasing sequences'
def MISSNODE(l):
    rows = []
    for s in MISS(l):
        row = []
        for i in range(len(l)):
            if l[i] in s:
                row.append(i)        
        rows.append(row)
    return rows
```

```python
'Finding Fulkerson dual family'
def dual(l):
    m = len(l)
    N =  MISSR(l)
    # initialize an empty list of rows for the augmented matrix
    rows = []
    # add rows corresponding to the constraints rho >= 0
    for i in range(1,len(l)+1):
        row = (m+1)*[0]
        row[i] = 1
        rows.append(row)   
    # add rows corresponding to the constraints N*rho >= 1
    for a in N:
        row = [-1] + a
        rows.append(row)   
    # create the polyhedron in cdd
    mat = cdd.Matrix(rows, number_type='fraction')
    mat.rep_type = cdd.RepType.INEQUALITY
    poly = cdd.Polyhedron(mat)
    ext = poly.get_generators()
    # list of dual objects
    dual = []
    # loop over extreme points and directions
    for i in range(ext.row_size):
    
        # skip extreme directions
        if ext[i][0] == 0:
            continue
        
        # add the vector representation of the dual object
        dual.append(ext[i][1:])
    print('Number of maximal increasing sequences=',len(MISSR(l)))
    print('Number of extremal points=',len(dual))
    return dual
```

```python
'graph formulation for a permutation'
def DPG(l):
    G = nx.DiGraph()
    for i in range(len(l)):
        G.add_node(i)
    for i in range(len(l)):
        for j in range(len(l)):
            if i < j:
                if l[i]<l[j]:
                    passed= True
                    for k in range(i,j):
                        if l[i]<l[k]<l[j]:
                            passed = False
                    if passed:
                        G.add_edge(i,j)
    pos = nx.spring_layout(G)
    labels = {u:(u+1)  for u in G.nodes}
    nx.draw(G, pos, node_size=200, node_color='blue', edge_color='black')
    nx.draw_networkx_labels(G, pos,labels=labels, font_size=10, font_color='yellow')
    S= []
    T= []
    for x in G.nodes():
        source = True
        sink = True
        for (u,v) in G.edges():
            if x == u:
                source = False
            if x == v:
                sink = False
        if source:
            T.append(x)
        if sink:
            S.append(x)
    
    return G, S, T
```

```python
'Add a source and a sink to the graph and generate list of all paths.'
def CPD(G,S,T):
    G.add_node('s')
    G.add_node('t')
    for u in S:
        G.add_edge('s',u)
    for u in T:
        G.add_edge(u,'t')
    path = nx. all_simple_paths(G, 's', 't')
    P = []
    for u in path:
        P.append(u[1:-1])
    return P
```

```python
np.random.seed(123)
l =   np.random.permutation(range(1,14))
print('Permutation=',l)
print('Number of maximal increasing sequences=',len(MISS(l)))
print('Lists of all maximal increasing sequences:',MISSNODE(l))

'Graph formulation'
G,S,T = DPG(l)
print('List of all paths:',CPD(G,S,T))
print('Number of paths:',len(CPD(G,S,T)))    
```

```python
def vertexpath(G,S,T):
    G.add_node('s')
    G.add_node('t')
    for u in S:
        G.add_edge('s',u)
    for u in T:
        G.add_edge(u,'t')
    path = nx. all_simple_paths(G, 's', 't')
    P = []
    for u in path:
        P.append(u[1:-1])
    m = len(G.nodes())
    rows = []
    for s in P:
        row = (m-2)*[0]
        for i in range(m-2):
            if i in s:
                row[i]=1        
        rows.append(row)
    return rows
```
