```python
def matrix(G):
    for i,(u,v) in enumerate(G.edges()):
        G[u][v]['enum'] = i
    n = len(G.nodes)
    N=[]
    for T in combinations(G.edges,n-1):
            H = nx.Graph(T)
            if nx.is_tree(H):   
                v = np.zeros(m)
                for (a,b) in T:
            
                    ind = G[a][b]['enum']
                    v[ind] = 1
                N.append(v)       
    return(np.array(N))

def modul(N,p,sigma):
    # number of rho variables
    n = N.shape[-1]
    # make sigma into a diagonal matrix
    S = np.diag(sigma)
    
    # set up optimization variables
    rho = cvx.Variable(n)
    
    # YOUR CODE HERE
    X= (S**(1./p)) @ rho
    obj = cvx.Minimize(cvx.pnorm(X,p)**p)
    cons = [rho>=0,N @ rho >=1 ]
    # solve the problem
    prob = cvx.Problem(obj,cons)
    prob.solve(solver='ECOS')
    return  np.array(obj.value).flatten(), np.array(rho.value).flatten()
```

```python
import itertools
import numpy as np
import networkx as nx

try:
    import sympy as sp
    HAVE_SYMPY = True
except Exception:
    HAVE_SYMPY = False


def tree_edge_incidence_matrix(G):
    """
    Build N âˆˆ {0,1}^{(#spanning trees) x (#edges)}
    where N[T, e] = 1 iff edge e appears in spanning tree T.
    """
    if not nx.is_connected(G):
        raise ValueError("G must be connected to have spanning trees.")

    # Fix a stable indexing for edges
    edges = sorted([tuple(sorted(e)) for e in G.edges()])
    m = len(edges)
    edge_index = {e: i for i, e in enumerate(edges)}

    rows = []
    # ðŸ”‘ correct function is enumerate_spanning_trees
    for T_edges in nx.algorithms.tree.enumerate_spanning_trees(G):
        row = np.zeros(m, dtype=int)
        for e in T_edges:
            e = tuple(sorted(e))
            row[edge_index[e]] = 1
        rows.append(row)

    if len(rows) == 0:
        return np.zeros((0, m), dtype=int), edges
    return np.vstack(rows), edges



def rank_over_reals(N, tol=None):
    """
    Numerical rank via SVD (NumPy). Optionally set a custom tolerance.
    """
    if tol is None:
        return np.linalg.matrix_rank(N)
    else:
        # manual SVD thresholding
        s = np.linalg.svd(N.astype(float), compute_uv=False)
        return int((s > tol).sum())


def rank_over_rationals(N):
    """
    Exact rank over the rationals via SymPy (if available).
    """
    if not HAVE_SYMPY:
        return None
    return sp.Matrix(N.tolist()).rank()


def test_one_graph(n=5, radius=0.8, seed=None, exact=False, verbose=True):
    """
    Build a random geometric graph, construct N, compute rank,
    and report full-rank status.
    """
    G = nx.random_geometric_graph(n, radius, seed=seed)
    # Make sure itâ€™s connected; if not, take the largest component
    if not nx.is_connected(G):
        G = G.subgraph(max(nx.connected_components(G), key=len)).copy()

    N, edges = tree_edge_incidence_matrix(G)
    trees, m = N.shape

    if exact and HAVE_SYMPY and trees * m <= 20000:  # small safety cap
        r = rank_over_rationals(N)
        method = "SymPy (exact)"
    else:
        r = rank_over_reals(N)
        method = "NumPy (numerical)"

    is_full = (r == min(trees, m))
    if verbose:
        print(f"n={G.number_of_nodes()}, m={m}, #trees={trees}")
        print("Shape (trees, edges):", N.shape)
        print(f"Rank of matrix N ({method}): {r}")
        print(f"Full rank? r == min(trees, edges) â†’ {is_full}")
        print("-" * 50)
    return is_full, r, N.shape


def sweep_trials(
    n_list=(5, 6, 7),
    radius_list=(0.6, 0.8, 1.0),
    trials=10,
    exact=False,
):
    """
    Run multiple random trials across sizes & radii and tally full-rank hits.
    """
    results = []
    for n in n_list:
        for r in radius_list:
            full_count = 0
            for t in range(trials):
                seed = 10_000 * n + 1_000 * int(r * 100) + t
                try:
                    is_full, rank_val, shape = test_one_graph(
                        n=n, radius=r, seed=seed, exact=exact, verbose=False
                    )
                except Exception as e:
                    is_full, rank_val, shape = (False, None, None)
                full_count += int(is_full)
            results.append((n, r, trials, full_count))
            print(
                f"n={n:2d}, r={r:.2f} â†’ full-rank {full_count}/{trials} trials"
            )
    return results


if __name__ == "__main__":
    # --- Single example (mirrors your parameters) ---
    print("Single test graph:")
    _ = test_one_graph(n=5, radius=0.8, seed=3810312, exact=True, verbose=True)

    # --- Batch experiments to probe the conjecture ---
    print("\nBatch sweep:")
    _ = sweep_trials(
        n_list=(5, 6, 7),
        radius_list=(0.6, 0.8, 1.0),
        trials=5,
        exact=False,  # set True for tiny graphs if you want exact rank
    )
```
